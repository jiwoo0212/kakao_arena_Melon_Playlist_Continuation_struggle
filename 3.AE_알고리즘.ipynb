```python
import pandas as pd
train_all= pd.read_json('data/train.json', encoding='utf-8', typ = 'frame')
#test= pd.read_json('test.json', encoding='utf-8', typ = 'frame')
train_all=train_all.sort_values(by='like_cnt', ascending=False)
#test = test.sort_values(by='like_cnt', ascending=False)
train=train_all[0:100]
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
a_tags=train_all['tags']
```

```python
eb_model = Word2Vec(a_tags,         # 리스트 형태의 데이터
                 sg=1,         # 0: CBOW, 1: Skip-gram
                 size=100,     # 벡터 크기
                 window=2,     # 고려할 앞뒤 폭(앞뒤 3단어)
                 min_count=0,  # 사용할 단어의 최소 빈도(0회 이하 단어 무시)
                 workers=4)    # 동시에 처리할 작업 수(코어 수와 비슷하게 설정)

# model.init_sims(replace=True) 필요없는 메모리 unload
```

```python
eb_model.save('word2vec.model')
```

```python
eb_model = Word2Vec.load('word2vec.model')
```

```python
NUM_WORDS, EMB_DIM = eb_model.wv.vectors.shape
```

# MODEL

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Embedding,Input,LSTM,Dense,Activation
from tensorflow.keras.layers import add
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Embedding,Input,LSTM,Dense,Activation,Concatenate,Reshape, Lambda
from tensorflow.keras.layers import add, Reshape
from tensorflow.keras.optimizers import SGD, RMSprop, Adam
from tensorflow.keras.backend import expand_dims
from tensorflow.keras.layers import LeakyReLU


def slice_tensor1(input):
    return input[:,0:10,:]

def slice_tensor2(input):
    return input[:,10:171,:]


##########################################

# Input 1: tags (5,10)
tag_input_a = Input(shape =(10,),name = 'tag_input')

# Input 1: Embedding layer
tag_input_b = Embedding(input_dim=NUM_WORDS, output_dim=EMB_DIM,trainable=False, input_length = 10,weights=[eb_model.wv.vectors])(tag_input_a)

##########################################

# Input 2: mel_spectrogram (5,200,100)
mel_input_a = Input(shape=(161,100), name='mel_input')

##########################################

# Concatenation layer here
x = Concatenate(axis = 1)([tag_input_b, mel_input_a])
x2 = Dense(100, activation='relu')(x)
x3 = Dense(100, activation='relu', name = "journey_embeddings")(x2)

#output = Activation("softmax")(x3)

##########################################

out1 = Lambda(slice_tensor1)(x3)
out2 = Lambda(slice_tensor2)(x3)

output1 = Dense(1, activation='relu', name='tags')(out1)
print(output1)

output2 = Dense(100, activation='relu', name='plyst_pca')(out2)
print(output2)

# output1 = Dense(1, activation=LeakyReLU(alpha = 0.1))(out1)
# output2 = Dense(100, activation=LeakyReLU(alpha = 0.1))(out2)
    
'''# An abitrary number of dense, hidden layers here
xt1 = Dense(20, activation='relu')(x3)
xt2 = Dense(10, activation='relu')(xt1)


xm = Dense(100, activation='relu')(x3)
xm2 = Reshape((210, 100))(xm)'''

#print(model.summary())
##########################################
adam = Adam(lr=0.001)
## Compile and fit the model
model = Model(inputs=[tag_input_a, mel_input_a], outputs=[output1, output2])
model.compile(loss = {'tags':'mse', 'plyst_pca':'mse'},optimizer = 'adam', metrics={'tags':'accuracy', 'plyst_pca':'accuracy'})
print(model.summary())
# np.random.seed(21)
```

```python
model.fit([tags, plyst_pca], [tags, plyst_pca], batch_size=1, validation_split=0.2, epochs=50)
```

```python
#Test Set Evaluation.
y_pred = model.predict([tags, plyst_pca], batch_size=1)
```
